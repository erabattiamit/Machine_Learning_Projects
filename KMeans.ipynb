```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
dir(iris)
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'])
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2)
```

```python
df['cluster']=km.fit_predict(df[['petal length (cm)','petal width (cm)']])
```

```python
df
```

```python
df0=df[df['cluster']==0]
df1=df[df['cluster']==1]
km.cluster_centers_
```

```python
plt.xlabel('length')
plt.ylabel('width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'])
plt.scatter(km.cluster_centers_[ : ,0],km.cluster_centers_[ : ,1])
```

```python
km.cluster_centers_
```

```python
from sklearn.preprocessing import MinMaxScaler
model=MinMaxScaler()
```

```python
model.fit(df[['petal length (cm)']])
```

```python
df['petal length (cm)1']=model.transform(df[['petal length (cm)']])
```

```python
model.fit(df[['petal width (cm)']])
```

```python
df['petal width (cm)1']=model.transform(df[['petal width (cm)']])
```

```python
model.fit(km.cluster_centers_)
```

```python
km1=model.transform(km.cluster_centers_)
```

```python
plt.scatter(df['petal length (cm)1'],df['petal width (cm)1'])
plt.scatter(km1[ : ,0],km1[ : ,1])
```

```python
sse=[]
k_range=range(1,10)
for k in k_range:
 mod1= KMeans(n_clusters=k)
 mod1.fit(df[['petal length (cm)','petal width (cm)']])
 sse.append(mod1.inertia_)
```

```python
plt.plot(sse)
```

```python
plt.plot(k_range,sse)
```
