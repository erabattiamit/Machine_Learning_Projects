```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/carprices.csv')
```

```python
df
```

```python
from pandas.core.reshape.encoding import get_dummies
dummies=get_dummies(df['Car Model'])
```

```python
dummies
```

```python
merged=pd.concat([df,dummies],axis=1)
merged
```

```python
final=merged.drop(['Mercedez Benz C class','Car Model'],axis=1)
```

```python
final
```

```python
x=final.drop(['Sell Price($)'],axis=1)
```

```python
y=final['Sell Price($)']
```

```python
model=LinearRegression()
model.fit(x,y)
model.predict([[52000,5,1,0]])
```

```python
model.score(x,y)
```
